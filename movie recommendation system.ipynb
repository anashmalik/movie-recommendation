```python
def con (obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
nds
```

```python
nds["genres"]=nds["genres"].apply(con)
```

```python
nds.head()
```

```python
nds["keywords"]=nds["keywords"].apply(con)
```

```python
def convertcast (obj):
    l=[]
    g=0
    for i in ast.literal_eval(obj):
        if g!=3:
            l.append(i["name"])
            g=g+1
        else :
            break
    return l
```

```python
nds["cast"]=nds["cast"].apply(convertcast)
```

```python
def concrew (obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i["name"])
    return l
```

```python
t=nds["crew"].apply(concrew)
t
```

```python
nds["crew"]=t
```

```python
nds.head(1)
```

```python
def cond (obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i)
    return l
```

```python
nds["overview"]=nds["overview"].apply(lambda x:x.split())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
def stem(te):
    l=[]
    for i in te.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
da['tags'].apply(stem)
```

```python
v=cv.fit_transform(da['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity as cs
```

```python
dismat=cs(v)
```

```python
dismat[0]
```

```python
def reco (movie):
    ind=nds[nds['title']==movie].index[0]
    dis=sorted(enumerate(dismat[ind]),reverse=True,key=lambda x:x[1])[1:6]
    re=[]
    for i in dis:
        re.append(nds.iloc[i[0]].title)
    return re
    
```

```python
reco("Batman Begins")
```
